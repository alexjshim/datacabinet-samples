If $ S $ is a set of generators and $B$ is a linearly independent set of vectors. $|S| > |B|$

## Basis Theorem 
Let $ V $ be a vector space. All bases of $ V $ have the same size.

This can be easily proved using the Morphing Lemma because if $B1$ and $B2$ are two generators of $ V $, then $|B1| < |B2| $ and $|B2| < |B1|$ using the Morphing Lemma. Which proves that $|B1| = |B2|$

## Superset basis Lemma

For any vector space $|V|$ and any linearly independent set of vectors $|A|$ , $|A|$ can be extended to form a basis of $|V|$ . It is easy to prove when the vector space is of form $F^D$ because the dimension is fixed at $|D|$ and the grow algorithm would definitely terminate at $D$.

## The dimension principle
if $V$ is a subspace of $W$ then, $dim V \le dim W$
if $dim V = dim W$ , then $ V = W $

This can be proved using the Superset basis Lemma

## Row rank = Column rank for a mtrix $A$

This can be proved by writing $ A = BU $ . B consists of a basis for the column vectors. There are r columns in B which is also the dimension of the column space. U thus has r rows. Each row of A can be written as a linear combination of the rows of U. This proves that the row space of A is of dimension less than r. This proves that row rank <= column rank. IF we consider $A^T$ , row rank >= column rank. The two together prove than row rank = column rank.

## Direct Sum($W$): 

If two vector spaces share only the zero vector, then we define direct sum of vector spaces $U$ and $V$ to be {u + v for u in $U$ and v in $V$}

This can be shown by writing any vector in the direct sum as a linear combination of the generators

* The union of the basis of $U$ and $V$ is a basis

## Dimension and Linear function

We will develop a criterion for whether a matrix is invertible. This will work for matrix as well.

### Linear function invertibility

Linear function $f: U-> W$ is invertible if dim Ker f = 0 and dim Im $f$ = dim $W$ = dim $U$ the origin space

### Formulation of a linear function
Let $f : V -> W$ be a linear function. Let us try to define a subfunction $f^* : V^* -> W^*$
$W*$ is the image of f which means $f^*$ is by definition onto. Let w1, w2 ..., wr be a basis of this image $W^*$. $f^*$ is defined as equal to f whereever it is defined. The preimage of w1, w2 ... wr be v1, v2 ... vr. Let $V^*$ be the span of v1, v2 ... vr 

$f^*$ is one to one . $Ker f^*$ is 0 because if $f^*(\alpha_1 * v_1 +\alpha_2 * v_2 ... \alpha_n * v_r) = 0$ , then $(\alpha_1 * w_1 +\alpha_2 * w_2 ... \alpha_n * w_r) = 0$ . but this is chosen to be the basis. Thus, proved by contradiction

$f^*$ is onto by definition

Also, $v_1, v_2 ... v_n$ form a basis for $V^*$. By formulation these span $V^*$ and it can be proven that they are linearly independent. If they were not linearly independent some non-trivial set of $\alpha$s would make $\alpha_1 * v_1 + \alpha_2 * v_2 ... \alpha_n * v_n = 0$ but then applying f on both sides, the same alphas can make all $wi's$ to be 0 as well. but $w_i's$ are zero by formulation.

### Kernel-Image Theorem
V = Direct_Sum(Ker f + $V^*$)

* Ker f and $V^*$ share only the zero vector 
* Any vector in v has $f(v) = w$. Because $f^*$ is onto there has to be another $v^* \in V^*$ such that $f^*(v^*) = w$. Also, $f(v^*) = w$. Thus $f(v) - f(v^*) = 0 . Thus, f(v - v*) = 0 . Thus, $(v-v*) \in Ker(f)$ . Thus v = v* + u where u is an element in Ker(f) 

**dim $Ker(f)$ + dim Im $f$ = dim $V$**

This follows directly from the fact that dim Im $f$ is the same as the dimension of $V^*$

## Linear function invertibility theorem

Let $f:V->W$ be a linear function. It is invertible if Ker(f) = 0(one to one) and dim V = dim W (onto). 

## Rank-Nullity Theorem

For a n column matrix A

rank A + nullity A = n

## Matrix Invertibility

Let A be an R * C matrix, A is invertible iff $|R|=|C|$ and the columns of A are linearly independent.

## Annihilator

$V^o$ = {$u \in F^n : u \dot v = 0$ for every vector $v \in V$}

dim V + dim $V^o$ = n

$(V^o)^o = V$
